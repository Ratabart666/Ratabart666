```python
import pandas as pd
from sympy import *
from sympy.printing.mathml import mathml
init_printing(use_unicode=True) # allow LaTeX printing

#Datos
print('------------------------------------------------')
datos=pd.read_csv('Datos 2.csv')
datos.index=['Ensayo 1','Ensayo 2']
print('Estos son los datos obtenidos en el laboratorio:')
print(datos)
print('Analizamos la precisión de los datos a través de la desviación estandar')
desviacionestandardatosseries=datos.std().round(3)
desviacionestandardatospandas=pd.DataFrame({'Desviación estandar':desviacionestandardatosseries})
print(desviacionestandardatospandas)


#Punto1
print('------------------------------------------------')
CalculosVaseries=datos['Vh(mL)']-datos['Vc(mL)']
CalculosValista=CalculosVaseries.tolist()
CalculosVapandas=pd.DataFrame({'Va(mL)':CalculosValista})
CalculosVapandas.index=('Ensayo1','Ensayo2')
print('Estos son los resultados para cada ensayo de Va:')
print(CalculosVapandas.round())
print('Haciendo análisis estadistico y expresando el resultado en función de la media y su incertidumbre tenemos:')
print(CalculosVapandas.describe().round())#count se refiere al número de datos, esta linea imprime:cuartiles,media y desviación estandar
desviacionestandar=CalculosVapandas.std().round().tolist()
media=CalculosVapandas.mean().round().tolist()
print(str('(')+str(media[0])+str('+-')+str(desviacionestandar[0])+str(')')+'mL')
#Punto2
print('------------------------------------------------')
Calculosfactorseries=(datos['Vh(mL)']-CalculosVaseries)/(datos['Th(K)']-datos['Ta(K)'])
Calculosfactorpandas=pd.DataFrame({'[(Vh-Va)/(Th-ta)](mL/K)':Calculosfactorseries})
Calculosfactorpandas.index=('Ensayo1','Ensayo2')
print('Estos son los resultado para cada ensayo de[(Vh-Va)/(Th-ta)]:')
print(Calculosfactorpandas.round(3))
print('Haciendo análisis estadistico y expresando el resultado en función de la media y su incertidumbre tenemos:')
print(Calculosfactorpandas.describe().round(2))#count se refiere al número de datos, esta linea imprime:cuartiles,media y desviación estandar
desviacionestandar=Calculosfactorpandas.std().round(3).tolist()
media=Calculosfactorpandas.mean().round(3).tolist()
print(str('(')+str(media[0])+str('+-')+str(desviacionestandar[0])+str(')')+'mL/k')
#punto3
print('------------------------------------------------')
pendienteseries=datos['Vh(mL)']/datos['Th(K)']
Calculospendientepandas=pd.DataFrame({'[Vh(mL)/Th(K)](mL/K)':pendienteseries})
Calculospendientepandas.index=('Ensayo1','Ensayo2')
print('Estos son los resultado para cada ensayo de[(Vh)/(Th)](mL/K):')
print(Calculospendientepandas.round(3))
print('Haciendo análisis estadistico y expresando el resultado en función de la media y su incertidumbre tenemos:')
print(Calculospendientepandas.describe().round(3))#count se refiere al número de datos, esta linea imprime:cuartiles,media y desviación estandar
desviacionestandar=Calculospendientepandas.std().round(3).tolist()
media=Calculospendientepandas.mean().round(3).tolist()
print(str('(')+str(media[0])+str('+-')+str(desviacionestandar[0])+str(')')+'mL/k')
print('------------------------------------------------')
print('Finalmente todos los datos quedan organizados de la siguiente manera:')
datos['Va(mL)']=CalculosValista
datos['[(Vh-Va)/(Th-ta)](mL/K)']=Calculosfactorseries
datos['[Vh(mL)/Th(K)](mL/K)']=pendienteseries
print(datos.round(3))



```
