# Análisis cuantitativo

## Teoría

### Datos teoría

```python
R1, R2, R3, R4, R5, R6 = 51, 51, 10, 22, 51, 51  # Ω
V_fuente = 10  # V
```

### Circuito teoría

### Resultados teoría

Plantiemos las ecuaciones correspondientes para hallar los valores predichos de la corriente, asumamos una dirección de corriente arbitraria.

La suma de las corrientes entrantes y salientes en cada nodo(c,e,h,f) debe ser 0, matemáticamente:

**En nodo c:** $$I_{1}=I_{2}+I_{3}$$

**En nodo e:** $$I_{3}=I_{4}+I_{5}$$

**En nodo h:** $$I_{2}+I_{5}=I_{6}$$

**En nodo f:** $$I_{7}=I_{4}+I_{6}$$

La suma de los aumentos y caidas de potencial en un camino cerrado debe ser 0, matemáticamente:

**Por la malla 1:** 
$$V_{f}-I_{1}R_{1}-I_{2}R_{3}-(I_{2}+I_{5})R_{6}=0$$

**Por la malla  2:**
$$V_{f}-I_{1}R_{1}-I_{3}R_{2}-(I_{3}-I_{5})R_{4}=0$$
$$V_{f}-I_{1}R_{1}-(I_{1}-I_{2})R_{2}-(I_{1}-I_{2}-I_{5})R_{4}=0$$

**Por la malla 3:** 
$$V_{f}-I_{1}R_{1}-I_{3}R_{2}-I_{5}R_{5}-(I_{2}+I_{5})R_{6}=0$$
$$V_{f}-I_{1}R_{1}-(I_{1}-I_{2})R_{2}-I_{5}R_{5}-(I_{2}+I_{5})R_{6}=0$$

Reacomodando tenemos las siguientes ecuaciones 3x3:


$$\left.
\begin{array}{rcl}
     I_{1}R_{1}+I_{2}(R_{3}+R_{6})+I_{5}R_{6}&=&V_{f}
  \\ I_{1}(R_{1}+R_{2}+R_{4})+I_{2}(-R_{2}-R_{4})+I_{5}(-R_{4})&=&V_{f}
  \\ I_{1}(R_{1}+R_{2})+I_{2}(R_{6}-R_{2})+I_{5}(R_{5}+R_{6})&=&V_{f}
\end{array}
\right\}
$$






```python
b = [[V_fuente, V_fuente, V_fuente]]
A = [[R1, R3+R6, R6], [R1+R2+R4, -R2-R4, -R4], [R1+R2, R6-R2, R5+R6]]
sol = Alg.producto(Alg.inversa(A), b)
print('El vector solución teórica (I1 A,I2 A,I5 A) es:')
Alg.print_matriz(sol)
I1teor, I2teor, I5teor = sol[0][0], sol[0][1], sol[0][2]
print('La corriente y la diferencia de potencial teórica en el resistor R1 es:',
      I1teor, 'A ', I1teor*R1, 'V')
print('La corriente y la diferencia de potencial teórica en el resistor R2 es:',
      I1teor-I2teor, 'A ', (I1teor-I2teor)*R2, 'V')
print('La corriente y la diferencia de potencial teórica en el resistor R3 es:',
      I2teor, 'A ', I2teor*R3, 'V')
print('La corriente y la diferencia de potencial teórica en el resistor R4 es:',
      I1teor-I2teor-I5teor, 'A ', (I1teor-I2teor-I5teor)*R4, 'V')
print('La corriente y la diferencia de potencial teórica en el resistor R5 es:',
      I5teor, 'A ', I5teor*R5, 'V')
print('La corriente y la diferencia de potencial teórica en el resistor R6 es:',
      I5teor+I2teor, 'A ', (I5teor+I2teor)*R6, 'V')
```

## Experimento

### Aclaración

Las corrientes se miden positivas si van en el sentido del dibujo teórico, si no se miden negativas.

De igual manera las diferencias de potencial se miden negativas si la corriente va en el sentido del dibujo teórico, si no se miden negativas.

- Si usted conecta N bombillos (resistencias) idénticos
a una batería (fuente). ¿Cómo se compara el brillo
(potencia) de cada bombillo a medida que se conectan
más y más de ellos al circuito, si se conectan i) en
serie, y ii) en paralelo?
¿En qué caso la batería durará más? Justifique.

**R//** 
1) En el caso en el que el circuito esta en serie, el voltaje total se divide en cada resistencia. Por lo que el brillo para esta acomodación se divide en cada bombillo y como cada uno de estos recibe un menor potencial su brillo es mas bajo en la medida en que se ponen más bombillos.

2) En el caso en el que el circuito esta en paralelo, la diferencia de potencial en cada resistencia sería la misma y por tanto todos los bombillos tendian la misma intensidad de brillo.

$$P=\frac{V^{2}}{R}$$

Ahora note lo siguiente. En el circuito en serie la diferencia de potecial de todos los resistores suma $V$ en cambio en el circuito en paralelo cada resistencia mide un potencial $V$. Así, la potencia total entregada en el circuito en serie es  menor que la  potencia total entregada en el circuito en paralelo y por lo tanto, la bateria se descarga más rápido en el circuito en en paralelo.
