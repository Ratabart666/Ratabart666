# Librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from prettytable import PrettyTable
from prettytable.colortable import ColorTable, Themes
from prettytable import MSWORD_FRIENDLY
from scipy import constants
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
plt.style.use('dark_background')
%matplotlib widget
```

# Datos

```python
mpesas=301.4/1000#kg
mrodillo=653.8/1000#kg
Drodillo=1.8/100#m
cerodillo=0.0923#cal/kg*°C
pi=constants.pi
g=constants.g
T0=20#°C
Ta=22#°C
Tfideal=Ta+(Ta-T0)#°C
datos=pd.read_csv('datos.txt',delimiter='    ')
datos=datos[(datos['Temperatura(°C)'] >=20) & (datos['Temperatura(°C)'] <=24)]
datos=datos.sort_values(by='Temperatura(°C)',ignore_index=True)
W=pi*datos['numero de vueltas']*mpesas*g*Drodillo#J
ΔU=mrodillo*cerodillo*mrodillo*(datos['numero de vueltas']-T0)#cal
x=PrettyTable()
x.set_style(MSWORD_FRIENDLY)
x.field_names=['masa rodillo(kg)','masa pesas(kg)','Diametro rodillo(m)','Temperatura inicial(°C)','Temperatura final experimental ideal(°C)']
x.add_row([mrodillo,mpesas,Drodillo,T0,Tfideal])
print('Estos son los datos experimentales:')
print('Se realizó limpieza de datos')
print(datos.to_markdown())
print(' ')
print('-'*100)
print(x)
```

```python
def func(x, m, b):
    return m*x+b

print('Hacemos una ajuste lineal W vs ΔU.')
print(' ')

ajuste,covarianza= curve_fit(func,ΔU,W)
desviacionestandar=np.sqrt(np.diag(covarianza))
m=round(ajuste[0],3)
b=round(ajuste[1],3)
Δm=round(desviacionestandar[0],3)
Δb=round(desviacionestandar[1],3)
nespteo=4.184#J/cal 
err=abs(nespteo-m)/nespteo
print('El ajuste W(ΔU)=mΔU+b,tiene como parámetros:')
print(' ')
print(str('m=(')+str(m)+str('+-')+str(Δm)+str(')J/K,'),str('b=(')+str(b)+str('+-')+str(Δb)+str(')J.'))
print('Gráficando W vs ΔU con el ajuste y comparando los puntos experimentales tenemos:')

fig,ax = plt.subplots() 

x0=min(ΔU)
xf=max(ΔU)
x=np.linspace(x0,xf,1000)

ax.scatter(ΔU,W,color='red')
ax.plot(x,(x*m)+b,color='blue')
ax.set_xlabel('ΔU (Cal)')  
ax.set_ylabel('W(J)')  
ax.set_title('W vs ΔU')


```

```python
print('Por ende nesp=(',m,'+-',Δm,')J/K')
print('El error porcentual es:',round(err,4)*100,'%')
```
