# Aplicación Integración de montercalo a $\int_{0}^{1}\frac{1}{t^2+1} \, dx$

```python
def funcion(x):
    return 1/(x**2+1)
def integral_montecarlo(funcion,a,b,numero_de_muestras):
    numeros_aleatorios=funcion(np.random.uniform(a,b,numero_de_muestras)).tolist()#números aleatorios evaluados en f
    integral=(1/numero_de_muestras)*Alg.sumatoria(numeros_aleatorios)*(b-a)#Integral
    return integral
```

```python
fig,ax=plt.subplots()
for i in np.arange(100,100100,100):
    ax.scatter(i,integral_montecarlo(funcion,0,1,i),color='red')
    ax.set_title('Integración metódo de montercarlo'+r'$\int_{0}^{1}\frac{1}{t^2+1}dx$')
    ax.set_ylabel('Valor integral')
    ax.set_xlabel('Valor de la muestra')
print('El valor más aproximado de la integral es: '+str(integral_montecarlo(funcion,0,1,100000)))
```

# Comentando los resultados

La resolución análitica de la integral es $\frac{\pi}{4}\approx0.785$. Es evidente que al aumentar el tamaño de la muestra, el resultado converge hacia el valor análitico. Estos nos permite comprobar que nuestro metódo de integración de montecarlo es correcto. Por otro lado cabe resaltar que para la generación de la muestra de números aleatorios, se tiene que usar una distribución uniforme para que el metódo funcione. Hemos realizado entonces nuestra práctica computacional de manera **existosa**.
