```python
class compleja:
    #Asignar los correspondientes valores realeas e imaginarios a las entradas.
    def __init__(self,Re,Im):
        self.Re=Re
        self.Im=Im
        
    def suma (self,Other):
        #Se definió la suma.
        re=self.Re+Other.Re
        im=self.Im+Other.Im
        
        
        #Se retornó el string como se lee normalmente.
        return (str(re)+" + "+str(im)+"i")
    
    def resta (self,Other):
        #Se asignó la resta, pero cabe aclarar que la funcion suma también resta.
        re=self.Re-Other.Re
        im=self.Im-Other.Im
        
        #Se retornó el string como se lee normalmente.
        return (str(re)+" - "+str(im)+"i")
    
    def multi (self,Other):
        #Se awsignó variables a los componentes de los números complejos.
        
        a=self.Re
        b=self.Im
        c=Other.Re
        d=Other.Im
        
        #Se asigno la multiplicación según la formula.
        
        re=(a*c)-(b*d)
        im=(a*d)+(b*c)
        
        #Se retornó el string como se lee normalmente.
        
        return (str(re)+" + "+str(im)+"i")
    
    def divi (self,Other):
        
        #Se awsignó variables a los componentes de los números complejos.
    
        a=self.Re
        b=self.Im
        c=Other.Re
        d=Other.Im
        
        #Se asigno la división según la formula.
        re=((a*c)+(b*d))/((c*c)+(d*d))
        im=((b*c)-(a*d))/((c*c)+(d*d))
        
        #Se retornó el string como se lee normalmente.
        return (str(re)+" + "+str(im)+"i")
    
    
    def norma (self):
        
        #Se asigno variables para solo uno de los números complejos.
        a=self.Re
        b=self.Im
        
        #Se aplicó la formula en la función.
        
        n=(((a*a)+(b*b))**(1/2))
    
        return n
```

```python
z1=compleja(1,1)#1+i
z2=compleja(2,-3)#2-3i

z1.suma(z2)
```
