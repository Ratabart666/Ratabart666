# Tabla de frecuencias variables

```python
def tabla_serie(serie):
    maximo=serie.max()
    minimo=serie.min()
    clases=round(1+np.log2(len(serie)))
    amplitud=(maximo-minimo)/clases
    x = PrettyTable()
    x = ColorTable(theme=Themes.OCEAN)
    x.field_names = ['Intervalo '+serie.name,'Frecuencia absoluta','Frecuencia porcentual(%)']
    Intervalo=''
    Frecuencia=''
    Frecuencia_por=''
    for i in np.arange(minimo,maximo-amplitud,amplitud):
        Intervalo='['+str(round(i,2))+','+str(round(i+amplitud,2))+')'
        Frecuencia=serie[(serie>=i) & (serie<=i+amplitud)].count()
        Frecuencia_por=round(Frecuencia/len(serie)*100,2)
        x.add_row([Intervalo,Frecuencia,Frecuencia_por])
    i=maximo-amplitud
    Intervalo='['+str(round(i,2))+','+str(round(i+amplitud,2))+']'
    Frecuencia=serie[(serie>=i) & (serie<=i+amplitud)].count()
    Frecuencia_por=round(Frecuencia/len(serie)*100,2)
    x.add_row([Intervalo,Frecuencia,Frecuencia_por])
    
    return x




def tabla_datos_agrupados_1(datos):
    'Dada algunos datos devuelve su información tabulada, cada columna corresponde a una variable'
    print('Tablas de frecuencia datos agrupados')
    print(' ')
    print('Nota: El número de datos por columna es:',len(datos))
    for i in datos:
        print(' ')
        print(100*'-')
        print(' ')
        print(tabla_serie(datos[i]))
    return ''
        
    


print(tabla_datos_agrupados_1(muestra.drop(['Dpto_residencia','Tipo_empleo','d_hijo6','hijos','inlf'],axis=1)))
        
        
    
    

```

# Probabilidad de que la esposa de edad promedio participe en el mercado laboral

```python
edad_promedio=round(muestra['edad_esposa'].mean())
print('La edad promedio de la esposa es: '+str(edad_promedio)+' años')
```

La probabilidad de A bajo la condición que ya ocurrió B es:

$$P(A|B)= \frac{P(A \cap B)}{P(B)}$$

Por lo tanto tenemos que la probabilidad de que este en el mercado laboral, dado que tiene una edad promedio es:
$$P=\frac{tenga \ edad \ promedio \ y \ este \ en \ el \ mercado  \ laboral}{probabilidad \ de \ que \ tenga \ edad \ promedio}$$



```python
def buscar_intervalo(edad,serie):
    maximo=serie.max()
    minimo=serie.min()
    clases=round(1+np.log2(len(serie)))
    amplitud=(maximo-minimo)/clases
    for i in np.arange(minimo,maximo,amplitud):
        if edad in range(round(i),round(i+amplitud)):
            return [i,i+amplitud]
```

```python
rango=buscar_intervalo(edad_promedio,muestra['edad_esposa']) 
print('El intervalo considerado para la edad promedio de',edad_promedio,' años es: ',rango)
print(100*'-')
P1=len(muestra[(muestra['edad_esposa'] >=rango[0]) & (muestra['edad_esposa'] <rango[1])])/len(muestra)*100
P2=len(muestra[(muestra['edad_esposa'] >=rango[0]) & (muestra['edad_esposa'] <rango[1]) & muestra['inlf']==1])/len(muestra)*100
P=round(P1/P2)
print('La probabilidad de que la esposa de edad promedio participe en el mercado laboral es ',P,'%')
```

# Probabilidad de que la esposa de edad promedio participe en el mercado laboral

**Como este evento es el evento complemento del anterior**, entonces la probabilidad es igual a $100\%-probabilidad \ evento \ anterior$

```python
print('La probabilidad de que la esposa de edad promedio no participe en el mercado laboral es ',100-P,'%')
```
