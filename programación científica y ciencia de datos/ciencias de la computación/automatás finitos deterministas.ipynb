# acepta todas las cadenas en binario con por lo menos 3 0's

```python
from automata.fa.dfa import DFA
dfa = DFA(
    states={'q0', 'q1', 'q2','q3','q4','q5'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': 'q2', '1': 'q1'},
        'q1': {'0': 'q2', '1': 'q1'},
        'q2': {'0': 'q4', '1': 'q3'},
        'q3': {'0': 'q4', '1': 'q3'},
        'q4': {'0': 'q5', '1': 'q4'},
        'q5': {'0': 'q5', '1': 'q5'}, 
    },
    initial_state='q0',
    final_states={'q5'}
)
#dfa.show_diagram(path='./dfa1.png')
```
