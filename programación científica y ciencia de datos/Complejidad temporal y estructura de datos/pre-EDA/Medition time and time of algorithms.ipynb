```python
import time#segundos
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

plt.style.use('dark_background')
z=[]
def tiempo(n):
    inicio = time.time()
    for i in range(n):
        z.append(i)
    fin = time.time()
    return(fin-inicio)
n=1000000

x=[]
y=[]
for i in range(10):
    x.append(i*n)
    y.append(tiempo(i*n))
plt.scatter(x,y)



    
```

```python
def func(x, m, b):
    return m*x+b


ajuste,covarianza= curve_fit(func,x,y)
desviacionestandar=np.sqrt(np.diag(covarianza))
m=ajuste[0]
b=ajuste[1]

fig,ax = plt.subplots() 

x0=x[0]
xf=x[-1]
z=np.linspace(x0,xf,1000)

ax.scatter(x,y,color='red')
ax.plot(z,z*m+b,color='blue')
ax.set_xlabel('n')  
ax.set_ylabel('t(s)')  
ax.set_title('t vs n')

print('El tiempo adicional por cada iteraci√≥n es:',m,'segundos/iteracion')

```
