# Análisis Cuantitativo

## Toma 1 y 2 datos experimentales

```python
#Datos para cada toma
R=1#ohms
MH2O=1#g
C=1#cal/g°C

#Toma 1
T01=1
Pvst_1=pd.read_csv('Toma 1.txt')
#Toma 2
T02=1
Pvst_2=pd.read_csv('Toma 2.txt')
#Equivalancia J y C
Eq_teor=4.186#J/cal

#Gráficos
fig,ax=plt.subplots()
ax.scatter(Pvst_1['t(s)'],Pvst_1['P(W)'],label='Datos 1')
ax.scatter(Pvst_2['t(s)'],Pvst_2['P(W)'],label='Datos 2')
ax.set_title('Datos experimentales')
ax.set_ylabel('Potencia (W)')
ax.set_xlabel('tiempo (s)')
plt.show()
```

## Energía eléctrica y térmica toma 1 y 2

```python
E_elec1=[]
E_elec2=[]

for i in range(len(Pvst_1['t(s)'])):
    E_elec1.append(Alg.integral_trapecio_funciondes(Pvst_1['t(s)'].tolist(),Pvst_1['P(W)'].tolist(),0,i))

for i in range(len(Pvst_2['t(s)'])):
    E_elec2.append(Alg.integral_trapecio_funciondes(Pvst_2['t(s)'].tolist(),Pvst_2['P(W)'].tolist(),0,i))

E_termica1=(MH2O*C*(Pvst_1['T(°C)']-T01)).tolist()
E_termica2=(MH2O*C*(Pvst_2['T(°C)']-T02)).tolist()
```

## Equivalente mecánica calor toma 1

```python
fig,ax=plt.subplots()
Alg.dibujo_reglineal(E_elec1,E_termica1,0,0,'Toma 1','Energía térmica(cal)','Energía Eléctrica(J)',ax,label='Toma 1')
Eq_exp1=Alg.regresionlineal(E_elec1,E_termica1)[0][0]
error1=round((abs(Eq_exp1-Eq_teor)/Eq_teor)*100)
print('El valor experimental para la equivalencia entre calorias y Joules es:(',Eq_exp1,
      ')J/cal Comparando con el \n valor teórico que es:', Eq_teor, 'J/cal el error fue', error1, '%')
print('-'*23)
```

## Equivalente mecánica calor toma 2

```python
fig,ax=plt.subplots()
Alg.dibujo_reglineal(E_elec2,E_termica2,0,0,'Toma 2','Energía térmica(cal)','Energía Eléctrica',ax,label='Toma 2')
Eq_exp2=Alg.regresionlineal(E_elec2,E_termica2)[0][0]
error2=round((abs(Eq_exp2-Eq_teor)/Eq_teor)*100)
print('El valor experimental para la equivalencia entre calorias y Joules es:(',Eq_exp2,
      ')J/cal Comparando con el \n valor teórico que es:', Eq_teor, 'J/cal el error fue', error2, '%')
print('-'*23)
```
