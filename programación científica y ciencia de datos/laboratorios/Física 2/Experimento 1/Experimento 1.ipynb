Fusión:

```python

mH=33#g (masa hielo)
Cc=16.73#cal/C (Capacidad calorifíca del calorimetro(datos del fabricante))
mA=36#g(masa agua)
Ca=1#cal/g*C(Calor especifíco del agua)
Ti=36.8#C(Temperatura inicial agua caliente)
Tf=0.0#C(Temperatura final en equilibrio)
Δm=0.1
#La temperatura inicial del hielo es 0C
Lfteor=79.7#cal/g*C
Lfexp=(1/mH)*((Cc+mA*Ca)*(Ti-Tf)-mH*Ca*Tf)
ΔLfexp=abs((Cc*(Tf-Ti))/(mA**2))
errorexpf=abs((Lfteor-Lfexp)/Lfteor)*100
print('El calor latente de fusión obtenido experimentalmente es:(',round(Lfexp),'+-',round(ΔLfexp,1),')cal/g',)
print('---------------------------------')
print('El error experimental asociado es:',round(errorexpf),'%')
```

Vaporización:

```python
mA=64.4#g
mE=60.9#g
Ce=0.186#cal/g*C
m1=0.0643#C/s (variación temperatura con el tiempo)(ajuste lineal)
m2=-0.0112#g/s(variación masa con el tiempo)(ajuste lineal)
Lvapexp=(mA*Ca+mE*Ce)*abs((m1/m2))#Cal/g
Lvap=539#Cal/g
errorexpv=abs((Lvapexp-Lvap)/Lvap)*100
print('El calor latente de vaporización obtenido experimentalmente es:(',round(Lvapexp),')cal/g',)
print('---------------------------------')
print('El error experimental asociado es:',round(errorexpv),'%')
```
