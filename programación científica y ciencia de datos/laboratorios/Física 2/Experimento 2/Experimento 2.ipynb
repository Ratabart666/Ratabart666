# Análisis Cuantitativo

```python
Cagua=1.00#Cal/g*C
magua=148.0#g
Tf=27.8#c
Tiagua=28.9#C
TiC=13.9#C
CCexp=magua*Cagua*abs((Tf-Tiagua)/(Tf-TiC))#Cal/g*C
CCteor=16.7#Cal/g*C
error=(abs(CCexp-CCteor)/CCteor)*100
print('La capacidad calorífica experimental del calorímetro es:',round(CCexp),'Cal/C')
print('El error(si suponemos real la capacidad calorífica del fabricante) es',round(error),'%')
CC=CCteor#Se utiliza el valor especificado por el fabricante
```
