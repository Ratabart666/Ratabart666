```python
datos=[datos1,datos2,datos3,datos4]
c=1
regresiones=[]
fig,ax=plt.subplots()
for i in datos:
    Alg.dibujo_reglineal(i["I(A)"].tolist(),(i["m(kg)"]*gravity).tolist(),0,0,'Datos experimentales','I(A)','F(N)',ax,'Datos'+str(c))
    regresiones.append(Alg.regresionlineal(i["I(A)"].tolist(),(i["m(kg)"]*g).tolist()))
    c+=1
```

Hallando el campo, tenemos:

$$F=(LB)I$$

Donde I es variable y LB son constantes, si sacamos la pendiente por regresi√≥n lineal nos queda:

$$m=LB$$
Lo anterior equivale:
$$B=\frac{m}{L}$$

```python
for i in range(len(regresiones)):
    print('Para los datos experimentales '+str(i+1)+' el campo magnetico es: '+str(round(regresiones[i][0][0]/L[i],2))+' T')
```
