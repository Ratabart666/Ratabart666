```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np
```

```python
#Datos
```

```python
datos1=pd.read_csv('efecto concentracion permanganato enfriado.csv')
datos2=pd.read_csv('efecto concentracion permanganato ambiente.csv')
datos3=pd.read_csv('Catalizador temperatura ambiente.csv')
print('Estos son los datos experimentales')
print(' ')
print(datos1)
print('')
T1=273.65#K
print('La temperatura de la reacción(sin catalizar) es:',T1,'k')
print('---------------------')
print(' ')
print(datos2)
T2=292.65#K
print(' ')
print('La temperatura de la reacción(sin catalizar) es:',T2,'k')
print('---------------------')
print(datos3)
print('')
T3=292.65#K
print('La temperatura de la reacción(catalizada) es:',T3,'k')
print('---------------------')


```

```python
#Gráficas
```

```python
fig,ax=plt.subplots()
ax.scatter(datos1['t(s)'],np.log((datos1['V KMnO4(L)(0.01M)']*0.01)/(datos1['V KMnO4(L)(0.01M)']+datos1['V agua(L)'])))
ax.set_ylabel('ln[KMnO4]')
ax.set_xlabel('t(s).')
ax.set_title('Cinética sin catalizador T='+str(T1)+'k'+' escala logaritmica')

def func(x,m,b):
    return m*x+b
ajuste,covarianza= curve_fit(func,datos1['t(s)'],np.log((datos1['V KMnO4(L)(0.01M)']*0.01)/(datos1['V KMnO4(L)(0.01M)']+datos1['V agua(L)'])))
m1=ajuste[0]
b1=ajuste[1]
x1=datos1['t(s)'].tolist()[0]
x2=datos1['t(s)'].tolist()[-1]
x=np.linspace(x1,x2,1000)
ax.plot(x,func(x,m1,b1),color='red')
print('Haciendo el ajuste tenemos que la constante de velocidad para T='+str(T1)+' k'+' es '+str(round(-m1,4)),'s**-1')
print('La concentración inicial es:',round(np.e**b1,2),'M')


```

```python
fig,ax=plt.subplots()
ax.scatter(datos1['t(s)'],(datos1['V KMnO4(L)(0.01M)']*0.01)/(datos1['V KMnO4(L)(0.01M)']+datos1['V agua(L)']))
ax.set_title('Cinética sin catalizador T='+str(T1)+'k'+' escala normal')
ax.set_ylabel('[KMnO4]')
ax.set_xlabel('t(s)')
ax.plot(x,np.e**func(x,m1,b1),color='red')
```

```python
fig,ax=plt.subplots()
ax.scatter(datos2['t(s)'],np.log((datos2['V KMnO4(L)(0.01M)']*0.01)/(datos2['V KMnO4(L)(0.01M)']+datos2['V agua(L)'])))
ax.set_ylabel('ln[KMnO4]')
ax.set_xlabel('t(s).')
ax.set_title('Cinética sin catalizador T='+str(T2)+'k'+' escala logaritmica')

def func(x,m,b):
    return m*x+b
ajuste,covarianza= curve_fit(func,datos2['t(s)'],np.log((datos2['V KMnO4(L)(0.01M)']*0.01)/(datos2['V KMnO4(L)(0.01M)']+datos2['V agua(L)'])))
m2=ajuste[0]
b2=ajuste[1]
x1=datos2['t(s)'].tolist()[0]
x2=datos2['t(s)'].tolist()[-1]
x=np.linspace(x1,x2,1000)
ax.plot(x,func(x,m2,b2),color='red')
print('Haciendo el ajuste tenemos que la constante de velocidad para T='+str(T2)+' k'+' es '+str(round(-m2,3)),'s**-1')
print('La concentración inicial es:',round(np.e**b2,2),'M')




```

```python
fig,ax=plt.subplots()
ax.scatter(datos2['t(s)'],(datos2['V KMnO4(L)(0.01M)']*0.01)/(datos2['V KMnO4(L)(0.01M)']+datos2['V agua(L)']))
ax.set_ylabel('[KMnO4]')
ax.set_xlabel('t(s)')
ax.set_title('Cinética sin catalizador T='+str(T2)+'k'+' escala normal')
ax.plot(x,np.e**func(x,m2,b2),color='red')

```

```python
fig,ax=plt.subplots()
ax.scatter(datos3['t(s)'],np.log((datos3['V KMnO4(L)(0.01M)']*0.01)/(datos3['V KMnO4(L)(0.01M)']+datos3['V agua(L)'])))
ax.set_ylabel('ln[KMnO4]')
ax.set_xlabel('t(s)')
ax.set_title('Cinética con catalizador T='+str(T3)+'k'+' escala logaritmica')

def func(x,m,b):
    return m*x+b
ajuste,covarianza= curve_fit(func,datos3['t(s)'],np.log((datos3['V KMnO4(L)(0.01M)']*0.01)/(datos3['V KMnO4(L)(0.01M)']+datos3['V agua(L)'])))
m3=ajuste[0]
b3=ajuste[1]
print('Haciendo el ajuste tenemos que la constante de velocidad para T='+str(T3)+' k'+' es '+str(round(-m3,2))+'s**-1')
print('La concentración inicial fue:',round(np.e**b3,2),'M')

x1=datos3['t(s)'].tolist()[0]
x2=datos3['t(s)'].tolist()[-1]
x=np.linspace(x1,x2,1000)
ax.plot(x,func(x,m3,b3),color='red')
```

```python
fig,ax=plt.subplots()
ax.scatter(datos3['t(s)'],(datos3['V KMnO4(L)(0.01M)']*0.01)/(datos3['V KMnO4(L)(0.01M)']+datos3['V agua(L)']))
ax.set_ylabel('[KMnO4]')
ax.set_xlabel('t(s)')
ax.set_title('Cinética con catalizador T='+str(T3)+'k'+' escala normal')
ax.plot(x,np.e**func(x,m3,b3),color='red')

```

```python
print('Comparando las constantes de velocidad sin y con catalizador,notamos que esta ultima es aproximadamente',round(m3/m2),'veces la constante   de la reacción sin catalizar')
```

```python
#Energía de activación
```

```python
R=8.314#J/K*mol
Ea=np.log(m1/m2)*R*T1*T2*(1/(T1-T2))#J
Ea=round(Ea/1000)#KJ
print('La energía de activación es',Ea,'kJ/mol')
```

```python
print('Con ecuaciones termoquimicas gráficamos la energía libre de gibbs')
ΔG1=-R*T1*np.log(-m1)
ΔG2=-R*T2*np.log(-m2)
ΔG3=-R*T3*np.log(-m3)
print(' ')
print('Reacción espontanea si ΔG<0 y no espontánea si ΔG>0 ')
print(round(ΔG1/1000),'KJ/mol',' T=',T1,)
print(round(ΔG2/1000),'KJ/mol',' T=',T2,)
print(' ')
print('Con catalizador:')
print(round(ΔG3/1000),'KJ/mol',' T=',T3,)
```
