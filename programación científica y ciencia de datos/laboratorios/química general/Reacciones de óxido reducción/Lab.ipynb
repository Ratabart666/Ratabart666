```python
import pandas as pd
datos11=pd.read_csv('datos individual Estandarización KMNO4.csv')
datos12=pd.read_csv('datos grupo Estandarización KMNO4.csv')
print('Los datos son los siguientes(Estandarización KMnO4)')
print(' ')
print('Réplicas')
print(datos11)
print(' ')
print('Datos grupos')
print(datos12)
M1=((datos11['masa Na2C2O4(g)']/134)*(2/5))/(datos11['Volumen KMnO4 0.1M(mL)']*(10**-3))
M2=((datos12['masa Na2C2O4(g)']/134)*(2/5))/(datos12['Volumen KMnO4 0.1M(mL)']*(10**-3))
print('-----------------------------------')
print('Resultados réplicas(concentación KMNO4):')
print(M1.round(2))
print(' ')
print('Media')
print(round(M1.mean(),2))
print(' ')
print('-----------------------------------')
print('Resultados generales (concentración KMNO4)(compañeros+ réplica individual) ')
print(M2.round(2))
print('Media')
print(' ')
print(round(M2.mean(),2))
print(' ')
print('Desviación estandar')
print(' ')
print(round(M2.std(),3))






```
