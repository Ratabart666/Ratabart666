```python
import pandas as pd 

titulacion1=pd.read_csv('Titulación NaOH.csv')
titulacion2=pd.read_csv('Titulación HCl.csv')
concentracionNaOH=round(((titulacion1['masa ftalato ácido de potasio(g)']/204.22)/(titulacion1['Volumen NaOH(L)'])).mean(),3)
titulacion2=titulacion2.rename(columns={'Volumen NaOH(L)':'Volumen NaOH(L) '+str(concentracionNaOH)+'M'})
rename='Volumen NaOH(L) '+str(concentracionNaOH)+'M'
print('Titulación-1')
print('                     ')
print(titulacion1)
print('       ' )
print('Media masa ftalato ácido de potasio(g) y Volumen NaOH(L):')
print('             ')
print(titulacion1.mean().round(3))
print('               ')
print('Desviación estándar masa ftalato ácido de potasio(g) y Volumen NaOH(L):')
print('          ')
std=titulacion1.std()
std.iloc[0]=round(std.iloc[0],2)
std.iloc[1]=round(std.iloc[1],4)
print(std)

print('------------------------------------------------------------------------')

molesconsumidas=(titulacion1['masa ftalato ácido de potasio(g)']/204.22).round(5)
concentracion=((titulacion1['masa ftalato ácido de potasio(g)']/204.22)/(titulacion1['Volumen NaOH(L)'])).round(3)
print('Moles consumidas NaOH')
print('                     ')
print(molesconsumidas)
print('                     ')
print('La media de moles consumidas:')
print(round(molesconsumidas.mean(),5))
print('                     ')
print('La desviación estándar de moles consumidas:')
print('                     ')
print(round(molesconsumidas.std(),5))

print('------------------------------------------------------------------------')

print('Concentración NaOH')
print('                     ')
print(concentracion)
print('                     ')
print('Media de la concentración NaOH:')
print('                     ')
print(round(concentracion.mean(),3))
print('                     ')
print('Desviación estándar de la concentración NaOH:')
print('                     ')
print(round(concentracion.std(),3))

print('------------------------------------------------------------------------')

print('Titulación-2')
print('                     ')
print(titulacion2)
print('                     ')
print('Media de Volumen HCl concentración desconocida(L) y '+rename)
print('                     ')
print(titulacion2.mean().round(4))
print('                     ')
print('Desviación estándar de Volumen HCl concentración desconocida(L) y '+rename)
print('                     ')
print(titulacion2.std().round(4))

print('------------------------------------------------------------------------')

concentracion2=(titulacion2[rename]*concentracionNaOH/titulacion2['Volumen HCl concentración desconocida(L)']).round(3)
print('Concentración HCL')
print('                     ')
print(concentracion2)
print('                     ')
print('La media de la concentración HCL:')
print('                     ')
print(round(concentracion2.mean(),3))
print('                     ')
print('La desviación estándar de la concentración HCL:')
print('                     ')
print(round(concentracion2.std(),3))
```
