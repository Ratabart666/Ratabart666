```python
%matplotlib notebook
from matplotlib import pyplot as plt
import numpy as np 
from matplotlib.animation import FuncAnimation 

fig = plt.figure() 
  
axis = plt.axes(xlim =(-7,7), 
                ylim =(-20,20))  
line, = axis.plot([], [], lw = 3) 
def init(): 
    line.set_data([], []) 
    return line, 
  
def animate(i): 
    x = np.linspace(-7,7, 100)
    z=np.linspace(-20,20,1000)
    # plots a sine graph 
    plt.plot(z*0,z,color='black')
    plt.plot(x,x*0,color='black')
    i=-3.5+(i/100)
    y =(i-2)*(1-2*(i+3)*np.sin(2*x))
    line.set_data(x, y) 
      
    return line, 
  
anim = FuncAnimation(fig, animate, 
                    init_func = init, 
                    frames = 550, 
                    interval =10, 
                    blit = True,repeat=True)
#si a pertenece al intervalo (-3.5,2.5) o a es igual a 2 la funciÃ³n no tiene nÃºmeros critÃ­cos
```
