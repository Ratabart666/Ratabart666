```python
#Datos continuos (Necesarios para el ajuste)
mc1 = sp.linspace(0, 10, 20)
mc2 = sp.linspace(0, 15, 20)
mc3 = sp.linspace(0, 14, 20)
mc4 = sp.linspace(0, 23, 40)
```

```python
#Constantes -(cm)-
D1= 222.5
D2= 218.0
D3= 211.5
D4= 224.7
d1= 0.009
d2= 2*d1
d3= 3*d1
d4= 4*d1
```

```python
def Y(m,a,b):
    return (a*m)+b
```

```python
params1, params_covariance = optimize.curve_fit(Y, m1, Ym1, p0=[2, 2])
params2, params_covariance = optimize.curve_fit(Y, m2, Ym2, p0=[2, 2])
params3, params_covariance = optimize.curve_fit(Y, m3, Ym3, p0=[2, 2])
params4, params_covariance = optimize.curve_fit(Y, m4, Ym4, p0=[2, 2])
```

```python
#Gr谩fica de m vs Ym para 1 hoja de papel
plt.figure(figsize=(8, 6))
plt.errorbar(m1, Ym1, yerr=0.05, xerr=0, fmt="o", mec='red', mfc='red', ecolor= "r", ms= "5", label = "Valores medidos")
plt.errorbar(mc1, Y(mc1, params1[0], params1[1]), color="g", fmt='--', label = "Ajuste")
plt.ylabel("Ym (cm) ", size = 'large') 
plt.yticks(size = 'large')
plt.xlabel("m", size = 'large') 
plt.xticks(X1, size = 'large')
plt.legend(loc="upper right")
plt.grid(True) 
plt.savefig("m vs Ym")
```

```python
x=m1
y=Ym1
N=9
a=params1[0]
b=params1[1]

#Incertidumbre de 位
a= mt.sqrt((sum((y-(a*x)-b)**2)*N)/((N-2)*(N*sum(x**2)-sum(x)**2)))
位= mt.sqrt((d1*a/D1)**2+(a*(0.0005)/D1)**2+((d1*a*(0.05)/(D1**2))**2))
print('位 = ' + str((a*d1/D1)*10**7) + ' nm') 
print('位 = ' + str(位*10**7) + ' nm') 
print('a = ' + str(a))
print('b = ' + str(b))
```
