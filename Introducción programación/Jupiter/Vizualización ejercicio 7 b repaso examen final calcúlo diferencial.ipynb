```python
%matplotlib notebook
from matplotlib import pyplot as plt
import numpy as np 
from matplotlib.animation import FuncAnimation 
  
# initializing a figure in 
# which the graph will be plotted 
fig = plt.figure() 
  
# marking the x-axis and y-axis 
axis = plt.axes(xlim =(-7,7), 
                ylim =(-20,20)) 
  
# initializing a line variable 
line, = axis.plot([], [], lw = 3) 
  
# data which the line will 
# contain (x, y) 
def init(): 
    line.set_data([], []) 
    return line, 
  
def animate(i): 
    x = np.linspace(-7,7, 100)
    z=np.linspace(-20,20,1000)
    # plots a sine graph 
    plt.plot(z*0,z,color='black')
    plt.plot(x,x*0,color='black')
    i=-3.5+(i/100)
    y =(i-2)*(1-2*(i+3)*np.sin(2*x))
    line.set_data(x, y) 
      
    return line, 
  
anim = FuncAnimation(fig, animate, 
                    init_func = init, 
                    frames = 550, 
                    interval =1, 
                    blit = True,repeat=False)
#Si a pertenece al intervalo (-3.5,-2.5) entonces f no tiene puntos crÃ­ticos,en el dibujo se muestra la grÃ¡fica de la derivada
#de f, se puede apreciar que en ningun momento la derivada se hace igual a cero en el intervalo (-3.5,-2.5)
```
