```python
%matplotlib notebook
from matplotlib import pyplot as plt
import numpy as np 
from matplotlib.animation import FuncAnimation 
  
# initializing a figure in 
# which the graph will be plotted 
fig = plt.figure() 
  
# marking the x-axis and y-axis 
axis = plt.axes(xlim =(-7,7), 
                ylim =(-1,1)) 
  
# initializing a line variable 
line, = axis.plot([], [], lw = 3) 
  
# data which the line will 
# contain (x, y) 
def init(): 
    line.set_data([], []) 
    return line, 
  
def animate(i): 
    a=2*np.pi+0.1-(np.pi/100)*i
    x = np.linspace(-7,7, 1000)
    z=np.linspace(-20,20,1000)
    # plots a sine graph 
    plt.plot(z*0,z,color='black')
    plt.plot(x,x*0,color='black')
    plt.plot(x,np.sin(x),color='red')
    m=((np.sin(np.pi)-np.sin(a)))/(np.pi-a)
    y=m*x-m*np.pi+np.sin(np.pi)
    line.set_data(x, y) 
      
    return line,
  
anim = FuncAnimation(fig, animate, 
                    init_func = init, 
                    frames = 100, 
                    interval = 1, 
                    blit = True,repeat=False)
anim.save('recta tangente en un punto.gif')
  
```
