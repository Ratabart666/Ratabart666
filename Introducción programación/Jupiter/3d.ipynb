```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


X = 3*np.linspace(-15,15,50)
Y = 3*np.linspace(-15,15,50)
X, Y = np.meshgrid(X,Y)

fig = plt.figure(figsize =(14, 9))
ax =fig.add_subplot(projection='3d')
Z=X*0+Y*0+10


ax.plot_surface(X, Y,Z)
ax.set_xlabel("x")

ax.set_ylabel("y")

ax.set_zlabel("z")

ax.plot([-40,40],[0,0],[0,0],color='black')
ax.plot([0,0],[-40,40],[0,0],color='black')
ax.plot([0,0],[0,0],[15,-15],color='black')

ax.scatter(np.linspace(-10,10,20),np.linspace(-10,10,20),np.linspace(-10,10,20),color='red')


plt.show()
#Faltan los subplots y las animaciones y experimentar transformaciones lineales(rotaciones sobre todo)
```
