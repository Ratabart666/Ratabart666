```python
import matplotlib.pyplot as plt
import numpy as np

pi=np.pi

x=np.arange(-6.28,6.28,0.01)#np.linspace(-6.28,6.28,100)
C,S=np.sin(x),np.cos(x)
plt.figure(figsize=(10,5),dpi=80)
plt.plot(x,C)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10,10)
plt.ylim(-2,2)
plt.xticks(np.arange((pi)*-2,(pi)*2,(np.pi)/2))
plt.yticks(np.arange(-1,2))
plt.plot(x, C, color="blue", linewidth=1, linestyle="-", label='coseno')
plt.plot(x, S, color="red",  linewidth=1, linestyle="-", label='seno')
plt.legend(loc='upper left')
plt.draw()
plt.fill

plt.plot(x,S,linewidth=1.0, linestyle="-")

z=np.linspace(-6.28,6.28)
w=z*0
plt.plot(z,w,color='black')

z=np.linspace(-1,1)
w=z*0
plt.plot(w,z,color='black')

   



```
